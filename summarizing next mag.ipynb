```python
import pandas as pd
import numpy as np
```

```python
sample = pd.read_csv('nextmag_tw_20160901', 
                     names = ['time','id1','location','name','user','type1','type2',
                              'type3','type4','ind','reading','device','from',
                'location2','view'],header = None, parse_dates = [0],
                    dtype = {'type4':str})
```

```python
sample
```

```python
sample.columns = ['time','id1','location','name','user','type1','type2','type3','type4','ind','reading','device','from',
                'location2','view']
# reset index as time
sample1 = sample.set_index('time')
```

```python
# 1. Counting number of lines every minute
sample2 = sample1.resample('1T', how = {'id1':'count'})
sample2
```

```python
# 2. Counting number of unique id1 every minute
sample3 = sample1.drop_duplicates(subset = {'id1'})
sample3 = sample3.resample('1T', how = {'id1':'count'})
sample3
```

```python
# 3. pivoting table by type1
sample4 = sample.pivot_table(index = {'time'}, columns = {'type1'},aggfunc = 'count', fill_value = 0)
sample4 = sample4.resample('1T', how = 'sum')
sample4
```

```python
# 4. pivoting table by type2
sample5 = sample.pivot_table(index = {'time'}, columns = {'type2'},aggfunc = 'count', fill_value = 0).resample('1T', how = 'sum')
sample5
```
